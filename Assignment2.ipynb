# Feature Transformation Techniques

```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv('/content/supershops.csv')
```

```python
df.head()
```

```python
df.shape # shape of the dataframe
```

```python
df.isnull().sum() # summary of checking null in each columns
```

```python
# drop null values
df.dropna(inplace=True) #inplace=True ensures that current dataframe df will be 
# changed with not null values removed
```

```python
df.shape
```

```python
# create 4 copies of the df to use in different techniques examples
df1 = df.copy()
df2 = df.copy()
df3 = df.copy()
df4 = df.copy()
```

# 1. **Normalization**: MinMaxScaler

```python
from sklearn.preprocessing import MinMaxScaler
m = MinMaxScaler()
```

```python
# directly apply normalization to 'Marketing Spend'column using fit_transform 
df1['Marketing Spend'] = m.fit_transform(df1[['Marketing Spend']])
df1.head()
```

```python
# now we'll first fit the 'Administration' column and then we'll transform it
admin_trnsfrm = m.fit(df1[['Administration']]) # first fit the column
df1['Administration'] = admin_trnsfrm.transform(df1[['Administration']]) # now transform it
# so we can either use fit_transform or we can do it in steps (first fit and then transform)
df1.head()
```

# 2. **Standardization**: StandardScaler

```python
from sklearn.preprocessing import StandardScaler
# https://scikit-learn.org/stable/modules/generated/sklearn.preprocessing.StandardScaler.html
ss = StandardScaler()

df2['Marketing Spend'] = ss.fit_transform(df2[['Marketing Spend']])
df2.head()
```

```python
df2['Transport'] = ss.fit_transform(df2[['Transport']])
df2.head()
```

```python
trnprt_trnsfrm = ss.fit(df2[['Transport']])
df2['Transport'] = trnprt_trnsfrm.transform(df2[['Transport']])
df2.head()
```

# **Max Absolute Scaler**: MaxAbsScaler

```python
from sklearn.preprocessing import MaxAbsScaler
# https://scikit-learn.org/stable/modules/generated/sklearn.preprocessing.MaxAbsScaler.html

abss = MaxAbsScaler()
```

```python
df3['Profit'] = abss.fit_transform(df3[['Profit']])
df3.head()
```

# **Robust Scaler**: RobustScaler

```python
from sklearn.preprocessing import RobustScaler
# https://scikit-learn.org/stable/modules/generated/sklearn.preprocessing.RobustScaler.html?highlight=robustscaler#sklearn.preprocessing.RobustScaler
```

```python
rs = RobustScaler()
df4['Marketing Spend'] = rs.fit_transform(df4[['Marketing Spend']])
df4.head()
```
